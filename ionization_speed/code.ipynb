### Скорость ступенчатой ионизации резонансно возбужденных атомов

В условиях коронарной модели заселенность некоторого ($i$-го) уровня атома вычисляется на основании
$$
    \frac{dn^*}{d\tau} = k_{0i} n_e n_n - n^* \sum_k\frac{1}{\tau_{ik}} = 0,
$$
где $n_e$, $n_n$ - концентрация электронов и нейтралов, соответсвенно.

Решение этого уравнения имеет вид
$$
    n^* = \frac{n_n n_e k_{0i}}{\sum_k\frac{1}{\tau_{ik}}}.
$$
Скорость ступенчатой ионизации за счет резонансного возбуждения атомов:
$$
    \boxed{ \left. \frac{dn_e}{d\tau} \right|_{st} = k_{st} \cdot n_e n^* = k_{st} \cdot \frac{n_n n^2_e k_{0i}}{\sum_{k}\frac{1}{\tau_{ik}}}, }
$$
$$
    k_{st} = \int^{\infty}_{v_i}  4\pi v^3 \sigma^*_i(v) f(v) dv,
$$
$$
    k_{0i} = \int^{\infty}_{v_i}  4\pi v^3 \sigma_{0i}(v) f(v) dv,
$$
где $\sigma^*_i$ - сечение ступенчатой ионизации возбужденного состояния атома, $\sigma_{0i}$ - сечение возбуждения резонансного состояния атома.

### Скорость ступенчатой ионизации как обратного процесса к тройной рекомбинации электронов и ионов
Тройная рекомбинация электронов и ионов:
$$
    A^+ + 2e \to A^* + e.
$$
Процесс ионизации, обратный тройной рекомбинации есть $A^* + e $:
$$
    \left. \frac{dn_e}{d\tau} \right|_{r} = k_{r} n_e n^*.
$$
Согласно принципу детального равновесия
$$
    k_r n_e n^* = \beta_3 n^2_e n_i \Rightarrow k_r = \beta_3\frac{n_e n_i}{n^*} = \frac{\lambda e^{10} }{m^{1/2}_e \left( T^{9/2}_e \right)} \cdot \frac{n_e n_i}{n^*},
$$
где $n_i$ - концентрация ионов, $n_e$ - концентрация электронов, $n^*$ - концентрация возбужденных атомов, $\beta_3$ - константа тройной рекомбинации, $\lambda = const$.

Из уравнения Саха (можно ли так поступать?)
$$
    n_i = A\frac{g_i}{g_n} \frac{n_n}{n_e} T^{3/2} \exp{\left( -\frac{U_i}{T_e} \right)},\ A = \left( \frac{2\pi m_e T_e}{h^2} \right)^{3/2}.
$$
У инертных газов $g_n = 1$, $g_i = 6$.

Скорость ступенчатой ионизации как обратного процесса к тройной рекомбинации электронов и ионов (для инетрных газов при $\lambda = 4$):
$$
    \boxed{ \left. \frac{dn_e}{d\tau} \right|_{r} = 24 n_e n_n m_e e^{10}  \cdot \left( \frac{2\pi}{h^2 T_e} \right)^{3/2}\exp{\left( -\frac{U_i}{T_e} \right)} }
$$
